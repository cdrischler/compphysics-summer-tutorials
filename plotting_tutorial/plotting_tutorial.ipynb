You can also plot more than one function in the same graph. For example

```
plot sin(x), x, tan(x)
```

You may want to specify ranges or limits for the x- and y-axes. A quick way to do so is in the [minimum:maximum] form before the function. The first square bracket will be specifying the x range, while the second specifies the y range. So, if there is only one square bracket, it will define the x-axis alone. Let's first leave off the x range, and then we can do both. Recall the previous plot command (using the [upp-arrow]) key and add the plot y range to be:

```
plot [] [-5:5] sin(x), x, tan(x)
```

You can see the y-axis now starts at -5 and ends at 5. Now let's define the same range for the x-axis:

```
plot [-5:5] [-5:5] sin(x), x, tan(x)
```

By default, gnuplot assumes that the independent variable for the plot command is $x$, but you can rename the independent variable. For example

```
plot [t=-5:5] [-5:5] sin(t), t, tan(t)
```

You can also define a function and refer to it each time you want to use it. Let's define $f(x)= a\,\sin(x)$ and plot it with different values of $a$. Type

```
f(x) = a*sin(x)
plot a=2, f(x), a=4, f(x), a=6, f(x)
```

You can see the legend by default has "a=..., f(x)" for each line. Let's edit the labels to only specify which value for "a" each line is. You can edit the legend by specifying the label of each function you are plotting. Recall the last command and add the `title` option right after each function

```
plot a=2, f(x) title "a=2", a=4, f(x) title "a=4", a=6, f(x) title "a=6"
```


### set and unset

The `set` and `unset` commands in Gnuplot allow you to configure and modify various plot settings and options.

The `set` command is used to set specific properties and parameters for the plot. It can be used to define axes labels, titles, line styles, colors, plot ranges, gridlines, and many other visual and formatting aspects. On the other hand, the `unset` command is used to remove or disable specific settings that have been previously set. 

The simplest use of `set` is to add a grid through. Try

```
set grid
replot
```

the `replot` command re-plots the last functions given. You will need to use it when setting/unsetting properties for your plot to see the updates.

Using `set` is another way you can define the ranges for your axes. Try

```
set xrange[0:9.4]
set yrange [-8:8]
replot
```

Now, try unsetting the x range by typing

```
unset xrange
replot
```

Another useful thing to do with `set` is to add labels on axes and a title to the plot. For example, try

```
set title "f(x) = a sin(x)"
set xlabel "x"
set ylabel "f(x)"
replot
```

You can change the font size for the title or labels as follows

```
set title font ",16"
set xlabel font ",14"
set ylabel font ",14"
replot
```

### Quick and simple

Let's generate a quick and simple plot. First, we define a range of x values, then use that range to plot $sin(x)$.


````python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

plt.plot(x,y)
plt.show()
````

Note that the more points you use, the smoother the curve will be.

<i> Try plotting the same graph with `x = np.linspace(0,2*np.pi,10)` instead. </i>

````python
x = np.linspace(0,2*np.pi,100)
f = np.sin(x)
g = np.cos(x)

plt.plot(x,f)
plt.plot(x,g)
plt.show()
````

### Plot attributes

Let's now customize our figures. There are so many attributes you can change in Matplotlib the way you exactly want. For the current tutorial, we are just going to go over the basics.


#### Labels and titles

You can add axis labels (x-axis and y-axis) and a title. Matplotlib also accepts TeX equation expressions in any text expression.

* `plt.title` is used to set a figure title
* `plt.xlabel` is used to set a label for the x-axis.
* `plt.ylabel` is used to set a label for the y - axis.


````python
plt.title('Exponential Decay')

x = np.linspace(0,10,100)

plt.plot(x, np.exp(-x))
plt.xlabel('x')
plt.ylabel('$e^{-x}$')
plt.show()
````

You can edit the size of the font and its color by adding `fontsize=` and `color=` as follows

````python
plt.title('Exponential Decay', fontsize = 16, color='red')

x = np.linspace(0,10,100)

plt.plot(x, np.exp(-x))
plt.xlabel('x', fontsize=14, color='black')
plt.ylabel('$e^{-x}$', fontsize = 14, color='black')
plt.show()
````

````python
x = np.linspace(0,10,100)

plt.plot(x, np.exp(-x))
plt.yscale("log")
plt.xscale("log")
plt.show()
````

#### Axes properties

You can customize axes properties, including the range of values, tick marks, tick labels, and the scale (linear or logarithmic).


````python
plt.title('Exponential Decay', fontsize = 16)

x = np.linspace(0,10,100)

plt.plot(x, np.exp(-x))
plt.xlabel('x', fontsize=14)
plt.ylabel('$e^{-x}$', fontsize = 14)

########## setting axes limits ###########
plt.xlim(0,3) 
plt.ylim(0,1) 


########## setting position of ticks ###########
plt.xticks(np.arange(0,3.5,0.5), fontsize=12) #position of xticks (does not include endpoint)
plt.yticks(np.arange(0,1.2,0.2), fontsize=12) #position of yticks (does not include endpoint)


plt.show()
````

#### Legends

For plots with multiple data series, you can add a legend to indicate the meaning of each series by adding a label to each one.

````python
plt.title('Exponential Decay', fontsize = 16)

x = np.linspace(0,10,100)

plt.xlabel('$x$', fontsize=14)
plt.ylabel('$e^{-\lambda x}$', fontsize = 14)

########## setting axes limits ###########
plt.xlim(0,3) 
plt.ylim(0,1) 

########## setting position of ticks ###########
plt.xticks(np.arange(0,3.5,0.5), fontsize=12) 
plt.yticks(np.arange(0,1.2,0.2), fontsize=12) 

########## defining the plots ###########
plt.plot(x, np.exp(-x), label='$\lambda = 1$')
plt.plot(x, np.exp(-2*x), label='$\lambda = 2$')

plt.legend() ## you have to include this line to make the legend visible

plt.show()
````

You can also plot different data series using a `for` loop.

````python
def f(c,x):
    return np.exp(-c*x)

plt.title('Exponential Decay', fontsize = 16)

x = np.linspace(0,10,1000)
l = np.arange(1,6,1)

plt.xlabel('$x$', fontsize=16)
plt.ylabel('$e^{-\lambda x}$', fontsize = 16)

########## setting axes limits ###########
plt.xlim(0,2) 
plt.ylim(0,1) 


########## setting position of ticks ###########
plt.xticks(np.arange(0,2.5,0.5), fontsize=12) 
plt.yticks(np.arange(0,1.2,0.2), fontsize=12) 

for i in l:
    plt.plot(x, np.exp(-i*x), label=f'$\lambda = {i}$')
    

plt.legend()
plt.show()
````

The legend can also be edited and tailored to how you want it to appear. For more details, check [legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html).

### Surface plots

Surface plots are convenient to capture complex relationships and to visualize functions of two variables or two-dimensional data. (see also [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html)). 

In the context of surface plots, `meshgrid` is a commonly used function that helps generate 2D grids from the given 1D arrays. This is particularly useful when creating surface plots because it allows you to evaluate a function over a grid of points in a 2D space, which is necessary to plot the 3D surface.

The meshgrid function takes two or more 1D arrays representing the x and y coordinates and returns two 2D arrays representing the x and y coordinates of the grid points. These grid points cover the entire 2D space formed by the combination of x and y values.

We will use a simple Gaussian as an example:

````python
import matplotlib.pyplot as plt
import numpy as np

# create a figure object
fig = plt.figure(dpi=100)

# plot intervals
xmin = -1; xmax = 1; ymin = -1; ymax = 1
s = 0.5

# set up a mesh of x,y values using numpy's meshgrid function
x = np.linspace(xmin,xmax,100)
y = np.linspace(ymin,ymax,100)

u, v = np.meshgrid(x, y)

Z = np.exp((-(u**2 + v**2)/(2*s**2)))

# decorate plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x,y)=\exp({-(x^2 + y^2)}~/~{2 \sigma ^2})$')

# create the surface plot
plt.imshow(Z, extent=[xmin,xmax,ymin,ymax])
plt.colorbar(shrink=0.5, aspect=10)
plt.show()
````
